```python
import pandas as pd
```

```python
df_applications=pd.read_csv('C:/users/ummsa/applications(2.0).csv')
df_industries=pd.read_csv('C:/users/ummsa/industries(2.0).csv')
```

```python
# 1.1 Removing duplicates by applicant_id
df_applications=df_applications.drop_duplicates('applicant_id')
display(df_applications)
```

```python
list(df_applications.columns)
```

```python
# Filling in missing values in External Rating and Education level
df_applications=df_applications.fillna({'External Rating' : 0, 'Education level' : 'Середня'})

#pd.set_option('display.max_rows', None)
#pd.reset_option('display.max_rows')

display(df_applications)
```

```python
# 3. Adding industry rankings
df_applications_merged=df_applications.merge(df_industries, on = 'Industry', how = 'left')
df_applications_merged
```

```python
# Determining the application ranking

# Conditions for calculating the ranking of each application
# age
R1=((df_applications_merged.Age>35)&(df_applications_merged.Age<55))*20

# Checking whether it is a holiday/weekend
df_applications_merged['Applied at']=pd.to_datetime(df_applications_merged['Applied at'], format='mixed')
R2=(df_applications_merged['Applied at'].dt.dayofweek.between(0,4))*20 

# Marital status
R3=(df_applications_merged['Marital status'] == 'Одружений')*20

# Locatin
R4=(df_applications_merged['Location'] == 'Київ чи область')*10

# Check External Rating
R5=(df_applications_merged['External Rating'] >= 7)*20
R6=(df_applications_merged['External Rating'] <= 2)*(-20)

# Data Frame generated
df_rating= pd.DataFrame({
    'R1': R1,
    'R2': R2,
    'R3': R3,
    'R4': R4,
    'R5': R5,
    'R6': R6,
    'R7': df_applications_merged.Score})
df_rating['total']=df_rating.sum(axis=1)
df_rating
```

```python
df_applications_merged ['Total rating']= df_rating['total']

# Replacing values below 0 and above 100
df_applications_merged.loc[df_applications_merged ['Total rating']<0, ['Total rating']]=0
df_applications_merged.loc[df_applications_merged ['Total rating']>100, ['Total rating']]=100

# Checking for missing values Amount and External Rating = 0
df_applications_merged.loc[(df_applications_merged['External Rating'] == 0)
                            &(df_applications_merged['Amount'].isnull()), ['Total rating']]=0

# 4.Finding row indices where in Total Rating 0 and delete them from DataFrame
row_totalrating_null = df_applications_merged.index[df_applications_merged ['Total rating'] == 0]
df_applications_final = df_applications_merged.drop(row_totalrating_null)
df_applications_final
```

```python
# 5. Average application rating per week

df_applications_final['year_week'] = (df_applications_final['Applied at'].dt.isocalendar().year.astype(str) + '-' +
    df_applications_final['Applied at'].dt.isocalendar().week.astype(str))
df_applications_week = (df_applications_final
                        .groupby(df_applications_final['year_week'])['Total rating']
                        .mean()
                        .reset_index(name='Average Rating')
                       )
df_applications_week['Average Rating']=round(df_applications_week['Average Rating'], 2)
df_applications_week
```
